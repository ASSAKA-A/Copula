```python
import numpy as np
import scipy.stats as ss
import pandas as pd
from pricing_model import geometric_brownian_motion
import matplotlib.pyplot as plt
```

```python
params= (100, 0.03, 0.02, 100, 1, 50)
```

```python
t,actif= geometric_brownian_motion(*params)
```

```python
plt.plot(t,actif)
plt.show()
```

```python
import seaborn as sns
sns.kdeplot(actif[-1], fill=True)
plt.axvline(np.median(actif[-1]), linestyle = "--", color ="red",label = "mediane")
plt.axvline(np.mean(actif[-1]), linestyle = "--", color ="green", label = "mean")
plt.legend()
```

```python
def portfolio(St,Mt):
    x = (St>Mt).astype(int)
    return x*S + (1-x)*Mt
```

```python
def placement_sans_risk(S0,rf):
    N=100
    t = np.linspace(0,1,N)
    return S0*np.exp(rf*t)
```

```python
Mt = placement_sans_risk(100,0.03)
```

```python
N,M = actif.shape
portefeuille = []
for i in range(M):
    x = (actif[i]>Mt[i]).astype(int)
    value = x*actif[i] + (1-x)*Mt[i]
    portefeuille.append(value)
```

```python
plt.plot(portefeuille)
plt.show()
```
